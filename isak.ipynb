```python
# Read data from files 
import pandas as pd

# Load the dataset
file_path = 'labeled_lyrics_cleaned.csv'
lyrics_data = pd.read_csv(file_path)

file_path = 'SpotifyFeatures.csv'
track_data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
lyrics_data.head()

```

```python
track_data.head()

lyrics_merge = pd.merge(lyrics_data, track_data, how='inner', left_on=['artist', 'song'], right_on=['artist_name', 'track_name'])

lyrics_merge.head()

```

```python
lyrics_merge['seq'] = lyrics_merge['seq'].replace('¶', ' ', regex=True)

# Add sentence length
lyrics_merge['lyrics_length'] = lyrics_merge['seq'].apply(lambda x: len(x.split()))

# Drop lyrics column
lyrics_merge.drop(columns=['seq'], inplace=True)

lyrics_merge.head()
```

```python
#print genres and their counts
print(lyrics_merge['genre'].value_counts())

# remove the rows with genre with less than 100 songs
lyrics_merge = lyrics_merge.groupby('genre').filter(lambda x: len(x) > 100)

print(lyrics_merge['genre'].value_counts())
```

```python
from sklearn.preprocessing import MinMaxScaler


numerical_cols = ['acousticness', 'danceability', 'duration_ms', 'energy', 
                  'instrumentalness', 'liveness', 'loudness', 'speechiness', 
                  'tempo', 'valence', 'lyrics_length']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Scale the numerical columns using MinMaxScaler
lyrics_merge[numerical_cols] = scaler.fit_transform(lyrics_merge[numerical_cols])
```

```python
import numpy as np

flop_or_bop_df = lyrics_merge

popularity_threshold = flop_or_bop_df['popularity'].median()
print(popularity_threshold)

flop_or_bop_df['flop_or_bop'] = np.where(
    flop_or_bop_df['popularity'] < popularity_threshold, 'flop', 'bop'
)
print(flop_or_bop_df['flop_or_bop'].value_counts())
```

```python
# Calculate popularity tertiles (33rd and 66th percentiles)
first_tertile = lyrics_merge['popularity'].quantile(0.33)
second_tertile = lyrics_merge['popularity'].quantile(0.66)

# Categorize into three groups: 'low', 'medium', 'high'
conditions = [
    lyrics_merge['popularity'] < first_tertile,
    lyrics_merge['popularity'] < second_tertile,
    lyrics_merge['popularity'] >= second_tertile
]

choices = ['low', 'medium', 'high']

lyrics_merge['popularity_category'] = np.select(conditions, choices)

lyrics_merge.head()
```

```python
# remove unwanted columns
lyrics_merge.drop(columns=['artist', 'song', 'col_0', 'label', "genre"], inplace=True)
```

```python
lyrics_merge = lyrics_merge.drop_duplicates(subset='track_id', keep='first')
```

```python
# Function to convert pitch to number
def pitch_to_number(pitch):
    pitch_map = {
        'C': 0,
        'C#': 1, 'Db': 1,
        'D': 2,
        'D#': 3, 'Eb': 3,
        'E': 4, 'Fb': 4,
        'E#': 5, 'F': 5,
        'F#': 6, 'Gb': 6,
        'G': 7,
        'G#': 8, 'Ab': 8,
        'A': 9,
        'A#': 10, 'Bb': 10,
        'B': 11, 'Cb': 11
    }
    return pitch_map.get(pitch.upper(), None)

# Replace the pitch column with its numeric representation
# Assuming the column name in your DataFrame that contains pitch values is 'pitch_column'
lyrics_merge['key'] = lyrics_merge['key'].apply(pitch_to_number)

# Compute recipe outputs from inputs
ready_for_training = lyrics_merge  # Now the DataFrame includes the transformed pitch values
```

```python
# to csv
ready_for_training.to_csv('ready_for_training.csv', index=False)
```
